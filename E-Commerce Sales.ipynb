### EXPLORATORY DATA ANALYSIS TO ANSWER QUESTIONS

```python
# Visualize the data to see the distribution of the purchasing prices of customers
highest_purchase_price = purchases_data["Purchase Price"].plot(kind="hist", bins=20, color="green", edgecolor='lightgreen', figsize=(10,6))
highest_purchase_price.set_xlabel("Purchase Price")
highest_purchase_price.set_ylabel("Frequency")
highest_purchase_price.set_title("Distribution of Purchase Prices")
plt.show()
```

```python
# Highest Purchase Price
highest_purchase_price = purchases_data[["Purchase Price"]].max()
highest_purchase_price
```

```python
# Lowest Purchase Prices
lowest_purchase_price = purchases_data[["Purchase Price"]].min()
lowest_purchase_price
```

```python
# Average Purchase Price
average_purchase_price = purchases_data[["Purchase Price"]].mean().round(2)
average_purchase_price
```

```python
# Do they buy more in the morning(AM) or Afternoon and Evening(PM)
morning_vs_afternoon = purchases_data.groupby("AM or PM")["Purchase Price"].mean()

morning_vs_afternoon.plot(kind='pie', figsize=(8, 8), autopct='%1.1f%%', startangle=90, colors=['green', 'lightgreen', 'lightcoral'])
plt.title('Time of the day VS Purchase Prices')
plt.ylabel('')
plt.show()
```

```python
# How many people purchased in the morning
morning_purchases = purchases_data[purchases_data["AM or PM"] == "AM"]
len(morning_purchases)
```

```python
# How many people purchased in the afternoon
morning_purchases = purchases_data[purchases_data["AM or PM"] == "PM"]
len(morning_purchases)
```

```python
# Morning and Afternoon Purchases
morning_afternoon_purchases = purchases_data["AM or PM"].value_counts()
morning_afternoon_purchases
```

```python
# Distribution of languages spoken
languages_spoken = purchases_data["Language"].value_counts()

plt.figure(figsize=(10, 6))
sns.countplot(data=purchases_data, x='Language', order=purchases_data['Language'].value_counts().index, palette='Greens')
plt.title('Distribution of Languages Spoken by Customers')
plt.xlabel('Language')
plt.ylabel('Count')
plt.xticks(rotation=45)

# Add annotations for each bar
for i, count in enumerate(languages_spoken):
    plt.text(i, count + 0.1, str(count), ha='center', va='bottom')

plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
# Number of customers employed VS Language

# Group the data by Language and count the number of occurrences of each job
job_types = purchases_data.groupby("Language")["Job"].count()

# Create a bar plot
plt.figure(figsize=(10, 6))
bar_plot = job_types.plot(kind="bar", color="green")
plt.title("Employed VS Language")
plt.xlabel("Language")
plt.ylabel("Number of customers employed")
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add gridlines to the y-axis

# Add annotations for each bar
for i, count in enumerate(job_types):
    plt.text(i, count + 0.1, str(count), ha='center', va='bottom')

plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
# Comparison of Purchase Behavior Across Credit Card Providers

# Group the data by 'CC Provider' and calculate the total purchase amount for each provider
purchase_amount_by_provider = purchases_data.groupby("CC Provider")["Purchase Price"].sum().sort_values(ascending=False)

# Create a bar plot
plt.figure(figsize=(10, 6))
purchase_amount_by_provider.plot(kind="bar", color="Green")
plt.title("Total Purchase Amount by Credit Card Provider")
plt.xlabel("Credit Card Provider")
plt.ylabel("Total Purchase Amount ($)")
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add gridlines to the y-axis
plt.tight_layout()  # Adjust layout to prevent clipping of labels

# Add annotations for each bar
for i, count in enumerate(purchase_amount_by_provider):
    plt.text(i, count + 0.1, str(count), ha='center', va='bottom')

plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
# How many people use credit cards that expire in 2020?
def counter():
    count = 0
    for date in purchases_data["CC Exp Date"]:
        if date.split("/")[1] == "20":
            count += 1
    print(count)
    
counter()
```

```python
# How many people use credit cards that expire in 2020?
len(purchases_data[purchases_data["CC Exp Date"].apply(lambda x:x[3:] == "20")])
```

```python
# Top 5 most popular email providers

# Create a for loop that takes in all of the mail providers from the for loop below
mail_provider = []

for email in purchases_data["Email"]:
    mail_provider.append(email.split("@")[1])

# Check the list
print(mail_provider)
```

```python
# Feature engineer a new column to take the mail providers
purchases_data["Mail Provider"] = mail_provider
```
