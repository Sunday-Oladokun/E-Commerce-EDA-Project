### EXPLORATORY DATA ANALYSIS TO ANSWER QUESTIONS

```python
# Highest Purchase Prices
highest_purchase_price = purchases_data[["Purchase Price"]].max()
highest_purchase_price
```

```python
# Lowest Purchase Prices
lowest_purchase_price = purchases_data[["Purchase Price"]].min()
lowest_purchase_price
```

```python
# Average Purchase Price
average_purchase_price = purchases_data[["Purchase Price"]].mean().round(2)
average_purchase_price
```

```python
# How many french speakers were there?
french_speakers = len(purchases_data[purchases_data["Language"] == "fr"])
french_speakers
```

```python
# Engineer job titles
engineers = purchases_data[purchases_data["Job"].str.contains("Engineer", case=False)] # Case is set to false to accommodate engineer in any case
engineers
```

```python
# How many engineers in all
len(engineers)
```

```python
# Find the email address of the person using this IP Address- 132.207.160.22
locate_mail_using_IP = purchases_data[purchases_data["IP Address"] == "132.207.160.22"]["Email"]
locate_mail_using_IP
```

```python
# How many people have masteercard as their card provider and made a purchase above 50?
mastercard_purchase_above_50 = purchases_data[(purchases_data["CC Provider"] == "Mastercard") & (purchases_data["Purchase Price"] >= 50)]
len(mastercard_purchase_above_50)
```

```python
# Find email of the person with the credit card number "4664825258997302"
locate_mail_using_credit_card_no = purchases_data[purchases_data["Credit Card"] == 4664825258997302]["Email"]
locate_mail_using_credit_card_no.head()
```

```python
# How many people purchased in the morning
morning_purchases = purchases_data[purchases_data["AM or PM"] == "AM"]
len(morning_purchases)
```

```python
# How many people purchased in the afternoon
morning_purchases = purchases_data[purchases_data["AM or PM"] == "PM"]
len(morning_purchases)
```

```python
# Morning and Afternoon Purchases
morning_afternoon_purchases = purchases_data["AM or PM"].value_counts()
morning_afternoon_purchases
```

```python
# Morning and Afternoon Purchases
purchases_data.groupby("AM or PM").count()
```

```python
# How many people use credit cards that expire in 2020?
def counter():
    count = 0
    for date in purchases_data["CC Exp Date"]:
        if date.split("/")[1] == "20":
            count += 1
    print(count)
    
counter()
```

```python
# How many people use credit cards that expire in 2020?
len(purchases_data[purchases_data["CC Exp Date"].apply(lambda x:x[3:] == "20")])
```

```python
# Top 5 most popular email providers

# Create a for loop that takes in all of the mail providers from the for loop below
mail_provider = []

for email in purchases_data["Email"]:
    mail_provider.append(email.split("@")[1])

# Check the list
print(mail_provider)
```

```python
# Feature engineer a new column to take the mail providers
purchases_data["Mail Provider"] = mail_provider
```

```python
purchases_data.head()
```

```python
# Top 5 email provider used
purchases_data["Mail Provider"].value_counts(sort=True).head()
```

```python
mail_providers = purchases_data.groupby("Mail Provider").size().reset_index(name="Count")
```

```python
mail_providers.sort_values("Count", ascending=False)
```

```python
# Visualize the top 5 email providers
purchases_data["Email"].apply(lambda x:x.split("@")[1]).value_counts().head().plot(kind="bar", rot=30)
```
